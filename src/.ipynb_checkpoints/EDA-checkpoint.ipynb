```python
data = pd.read_csv('../data/X_3.csv')
```

```python
fig, ax = plt.subplots()
ax.scatter(data['Gunning Fog Score'][data['Better than VGT?'] ==1 ],data['Ticker Data'][data['Better than VGT?'] ==1 ], label = 'Above VGT' )
ax.scatter(data['Gunning Fog Score'][data['Better than VGT?'] ==0 ],data['Ticker Data'][data['Better than VGT?'] ==0 ], label = 'Below VGT')
ax.set_ylabel('Stock Growth')
ax.set_xlabel('Gunning Fog Score')
ax.legend()
ax.set_title('Gunning Fog Score')
```

```python
fig, ax = plt.subplots()
ax.scatter(data['Felsch Kincaid Grade'][data['Better than VGT?'] ==1 ],data['Ticker Data'][data['Better than VGT?'] ==1 ], label = 'Above VGT' )
ax.scatter(data['Felsch Kincaid Grade'][data['Better than VGT?'] ==0 ],data['Ticker Data'][data['Better than VGT?'] ==0 ], label = 'Below VGT')
ax.set_ylabel('Stock Growth')
ax.set_xlabel('Felsch Kincaid Grade')
ax.legend()
ax.set_title('Felsch Kincaid Grade')
```

```python
data_lr = data.drop(['Transcript', 'Date', 'Ticker','Unnamed: 0.1','Unnamed: 0' ,'ticker_data','VGT'], axis=1)
data_lr = data_lr.dropna()
data_lr
```

```python
y = data_lr.pop('Better than VGT?')
X = data_lr
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y)
LR = LogisticRegression()
LR.fit(X_train,y_train)

print(LR.intercept_)
LR.coef_
```

```python
y = LR.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(LR.score(X_test, y_test)))
print(y)
print(y_test)
```
